# Customer EDA 

```python
# Importing necessary libraries for EDA
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Load the dataset
file_path = 'Customers.csv'
df = pd.read_csv(file_path)
```

```python
# Display the first few rows of the dataset to understand its structure
print(df.head())
```

```python
# Display basic information about the dataset
df.info()
```

```python
# Check for missing values
print("Missing values per column:")
print(df.isnull().sum())
```

```python
# Summary statistics of the dataset
print("Summary statistics:")
print(df.describe(include='all'))
```

```python
# Further EDA: Analyzing the distribution of regions and signup dates
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Convert SignupDate to datetime for analysis
df['SignupDate'] = pd.to_datetime(df['SignupDate'])
```

```python
# Distribution of customers by region
region_counts = df['Region'].value_counts()
plt.figure(figsize=(8, 5))
sns.barplot(x=region_counts.index, y=region_counts.values, palette='viridis')
plt.title('Distribution of Customers by Region')
plt.xlabel('Region')
plt.ylabel('Number of Customers')
plt.show()
```

```python
# Trend of signups over time
plt.figure(figsize=(10, 6))
df['SignupDate'].dt.to_period('M').value_counts().sort_index().plot(kind='line', marker='o', color='blue')
plt.title('Customer Signups Over Time')
plt.xlabel('Signup Month')
plt.ylabel('Number of Signups')
plt.grid(True)
plt.show()
```

```python
# Top 10 customers by name frequency
top_customers = df['CustomerName'].value_counts().head(10)
plt.figure(figsize=(8, 5))
sns.barplot(x=top_customers.values, y=top_customers.index, palette='coolwarm')
plt.title('Top 10 Customers by Name Frequency')
plt.xlabel('Frequency')
plt.ylabel('Customer Name')
plt.show()

print("EDA visualizations completed.")
```

```python
# Deriving business insights based on the EDA

# Insight 1: Regional distribution
region_distribution = df['Region'].value_counts(normalize=True) * 100
print("Regional distribution (%):")
print(region_distribution)
```

```python
# Insight 2: Signup trends over time
signup_trend = df['SignupDate'].dt.to_period('M').value_counts().sort_index()
print("Signup trends over time:")
print(signup_trend)
```

```python
# Insight 3: Most frequent customer names
most_frequent_names = df['CustomerName'].value_counts().head(10)
print("Most frequent customer names:")
print(most_frequent_names)
```

```python
# Insight 4: Yearly signup trends
yearly_signups = df['SignupDate'].dt.year.value_counts().sort_index()
print("Yearly signup trends:")
print(yearly_signups)
```

```python
# Insight 5: Unique customers and repeat patterns
unique_customers = df['CustomerID'].nunique()
total_customers = len(df)
repeat_customers = total_customers - unique_customers
print("Unique customers:", unique_customers)
print("Repeat customers:", repeat_customers)
```

# Product EDA

```python
# Read the dataset
df = pd.read_csv('Products.csv')
```

```python
# Display basic information about the dataset
print("Dataset Info:")
print(df.info())
```

```python
print("\
First few rows:")
print(df.head())
```

```python
print("\
Basic statistics:")
print(df.describe())
```

```python
print("\
Missing values:")
print(df.isnull().sum())
```

```python
print("\
Unique categories:")
print(df['Category'].value_counts())
```

```python
# Set style for better visualizations
plt.style.use('seaborn-v0_8')
```

```python
# 1. Price Distribution by Category
plt.figure(figsize=(12, 6))
sns.boxplot(x='Category', y='Price', data=df)
plt.title('Price Distribution by Category')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

INSIGHT - Books show the highest price variability, ranging from 33.59 dollars to 497.76 dollars, suggesting diverse product offerings from budget to premium titles.
Electronics maintain a consistent mid-to-high price range, indicating a focus on quality technology products.
Home Decor has the lowest average price point (235.02 dollars) but maintains a healthy price spread, appealing to various customer segments.

```python
# 2. Average Price by Category
plt.figure(figsize=(10, 6))
df.groupby('Category')['Price'].mean().plot(kind='bar')
plt.title('Average Price by Category')
plt.ylabel('Average Price')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

INSIGHT - The product catalog is well-balanced across categories with Books and Electronics having 26 products each, Clothing with 25 products, and Home Decor with 23 products, showing a relatively even distribution of inventory.

```python
# 3. Price Distribution (Overall)
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Price', bins=20)
plt.title('Price Distribution')
plt.xlabel('Price')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

#### Price Distribution Analysis
INSIGHT - The overall price distribution shows a bimodal pattern, with clusters around 150 dollars and 400 dollars, suggesting distinct market segments for budget and premium products.
There's a significant opportunity in the mid-range price segment (250-350 dollars) where product offerings are relatively fewer.

```python
# Calculate some additional metrics
print("\
Category-wise Statistics:")
print(df.groupby('Category').agg({
    'Price': ['count', 'mean', 'min', 'max']
}).round(2))
```

#### Category Performance
INSIGHT - Books category has the highest average price (292.21 dollars), followed by Electronics (275.35 dollars), indicating these categories drive higher revenue per unit.
Home Decor, despite having the lowest average price, maintains a healthy minimum price of 41.18 dollars, ensuring baseline profitability.

#### Strategic Pricing Insights
INSIGHT - The wide price range across all categories (16.08 dollars to 497.76 dollars) indicates a comprehensive market coverage strategy.
Each category has premium products priced above 450 dollars, showing consistent high-end market positioning across the product line.

# Transaction EDA

```python
from datetime import datetime
```

```python
# Read the data
df = pd.read_csv('Transactions.csv')
```

```python
# Display basic information
print("Dataset Overview:")
print("\
First few rows:")
print(df.head())
```

```python
print("\
Dataset Info:")
print(df.info())
```

```python
print("\
Basic Statistics:")
print(df.describe())
```

```python
# Checking for missing values and duplicates
print("Missing Values:")
print(df.isnull().sum())
```

```python
print("\
Duplicate Rows:")
print(df.duplicated().sum())
```

```python
# Converting TransactionDate to datetime format
df['TransactionDate'] = pd.to_datetime(df['TransactionDate'])
```

```python
# Extracting additional date-related features
df['Year'] = df['TransactionDate'].dt.year
df['Month'] = df['TransactionDate'].dt.month
df['Day'] = df['TransactionDate'].dt.day
df['Weekday'] = df['TransactionDate'].dt.day_name()

print("\
Date-related features added.")
```

```python
# Sales Analysis
# Monthly sales trend
monthly_sales = df.groupby(['Year', 'Month'])['TotalValue'].sum().reset_index()
monthly_sales['Date'] = pd.to_datetime(monthly_sales[['Year', 'Month']].assign(DAY=1))
monthly_sales = monthly_sales.sort_values('Date')
```

```python
plt.figure(figsize=(12, 6))
plt.plot(monthly_sales['Date'], monthly_sales['TotalValue'], marker='o')
plt.title('Monthly Sales Trend')
plt.xlabel('Date')
plt.ylabel('Total Sales Value')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

INSIGHT - The monthly sales trend visualization, shows distinct patterns in revenue generation throughout the year. There are noticeable peaks and troughs in sales, indicating seasonality in purchasing behavior that can be leveraged for inventory planning and marketing campaigns.

```python
# Top selling products
top_products = df.groupby('ProductID').agg({
    'Quantity': 'sum',
    'TotalValue': 'sum'
}).sort_values('TotalValue', ascending=False).head(10)

print("\
Top 10 Products by Sales Value:")
print(top_products)
```

INSIGHT - Product Performance Analysis of top-selling products reveals that product P029 is the highest revenue generator with total sales of 19,513 dollars , followed by P079 and P048. As shown in table, these top products consistently maintain high sales volumes, suggesting strong market demand and potential for inventory optimization.

```python
# Customer Analysis
customer_stats = df.groupby('CustomerID').agg({
    'TransactionID': 'count',
    'Quantity': 'sum',
    'TotalValue': 'sum'
}).rename(columns={'TransactionID': 'Number_of_Transactions'})

customer_stats['Average_Transaction_Value'] = customer_stats['TotalValue'] / customer_stats['Number_of_Transactions']
```

INSIGHT - The customer behavior analysis indicates that the average customer makes approximately 5 transactions, with some customers making up to 11 purchases. This suggests an opportunity for customer retention strategies, particularly since the top quartile of customers makes 6 or more purchases.

```python
# Distribution of transaction values
plt.figure(figsize=(10, 6))
sns.histplot(df['TotalValue'], bins=50)
plt.title('Distribution of Transaction Values')
plt.xlabel('Transaction Value')
plt.ylabel('Count')
plt.show()
```

```python
print("\
Customer Purchase Behavior Summary:")
print(customer_stats.describe())
```

INSIGHT - Transaction Value Distribution The distribution of transaction values shows a wide range, with most transactions clustering around certain price points. From Table, we can see that the average transaction value is around 687 dollars, with values ranging from 82 dollars to 1,323 dollars, indicating diverse product pricing strategies and customer spending patterns.
